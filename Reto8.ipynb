```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import urllib.request
```

```python
url= "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS9v65dMTIo7c1vHzsEmFBLuwSJyDkphfK50g&s"
Herida = np.array(Image.open(urllib.request.urlopen(url))) #Punto 1 
HeridaArray = np.array(Herida)
np.shape(HeridaArray)
plt.imshow(HeridaArray)
```
